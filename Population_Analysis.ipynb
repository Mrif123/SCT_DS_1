***Importing the Libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

***Loading the Dataset***



```python
df=pd.read_csv('/content/India_Population_By_Age_Group_2022.csv')
print(df.head())
print(df.shape)
print(df.columns)
```

***Bar Plot***

```python
# Set the style
sns.set_style("whitegrid")

# Create the bar plot
plt.figure(figsize=(10, 6)) # Adjust figure size as needed
sns.barplot(x='Age Group', y='Population (Millions)', data=df, palette='viridis') # Use Age Group and Population columns

# Add titles and labels
plt.title('Population Distribution by Age Group in India (2022)')
plt.xlabel('Age Group')
plt.ylabel('Population(Millions)')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

# Display the plot
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

***Histogram(Histplot)***

```python
# Set the style
sns.set_style("whitegrid")

# Create the histogram
plt.figure(figsize=(10, 6)) # Adjust figure size as needed
sns.histplot(data=df, x='Population (Millions)', kde=True) # Use the numerical column

# Add titles and labels
plt.title('Distribution of Population (Millions) in India (2022)')
plt.xlabel('Population (Millions)')
plt.ylabel('Age Group')

# Display the plot
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

***Line plot***

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age Group', y='Population (Millions)', data=df)
plt.xticks(rotation=45, ha='right')
plt.title('Population Trend Over Time')
plt.xlabel('Age Group')
plt.ylabel('Population (Millions)')
plt.tight_layout()
plt.show()
```

***Scatterplot***

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age Group', y='Population (Millions)', data=df)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Age Group')
plt.ylabel('Population (Millions)')
plt.title('Population by Age Group')
plt.show()
```

***Voilinplot***

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Age Group', y='Population (Millions)', data=df)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Age Group')
plt.ylabel('Population (Millions)')
```
